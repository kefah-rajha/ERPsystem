```python
import pandas as pd
from faker import Faker

users = pd.read_csv('New folder/user.csv')
companies = pd.read_csv('New folder/mycompanies.csv')

fake = Faker()

newUsers=users.drop(columns=['Subscription Plan','Renewal Status','Customer Support Interactions','Feedback/Ratings','Membership End Date'])
newUsers.rename(columns={'Name': 'name','Email Address' :'email','Date of Birth': 'Brithday','Membership Start Date':'createdAt'}, inplace=True)
newUsers['role'] = 'Customer'
newUsers['phone'] = [fake.numerify(text='+1 ### ### ####') for _ in range(len(newUsers))]
newUsers['city'] = [fake.city() for _ in range(len(newUsers))]
newUsers['street'] = [fake.street_address() for _ in range(len(newUsers))]  # شارع
newUsers['postcode'] = [fake.postcode() for _ in range(len(newUsers))] 
print(newUsers.columns)
print(newUsers)
newUsers.to_csv('Users.csv', index=False)
```
